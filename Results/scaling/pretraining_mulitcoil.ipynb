```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('../styles_and_utils/graphstyle.mplstyle')

import sys
sys.path.insert(1, '../')
from styles_and_utils import *
```

```python
w, h = plt.rcParams["figure.figsize"]
plt.rcParams.update({'font.size': 21})
fig, ax = plt.subplots(1, 1, figsize=(1*w, 1.3*h))

""" Model styles """
vitl = ViT('l', useline=True)
untl = Unet('l', useline=True)
vitm = ViT('m', useline=True)
untm = Unet('m', useline=True)

x = [100, 2000, 35000]
vnet = [0.891, 0.909, 0.918]
ptvit = [0.903, 0.907, 0.911]
ptunt = [0.891, 0.902, 0.907]
unt = [0.885, 0.898, 0.907]
vit = [0.851, 0.877, 0.908]

ax.plot(x, ptvit, '*--', markersize=vitm.ms, color=vitl.color, label='PT-'+vitl.label)
ax.plot(x, vit, vitl.marker, markersize=vitm.ms, color=vitl.color, label=vitl.label)
ax.plot(x, ptunt, '*--', markersize=untm.ms, alpha=untl.oc, color=untl.color, label='PT-'+untl.label)
ax.plot(x, unt, untl.marker, markersize=untm.ms, alpha=untl.oc, color=untl.color, label=untl.label)
ax.plot(x, vnet, 'p--', color='#8c2614', markersize=vitm.ms,label='E2E-VarNet')
ax.set_xscale('log')
ax.set_xticks(x)
ax.set_xticklabels(['100', '1k', '35k'])
ax.legend()
ax.set_xlabel('Training data')
ax.set_ylabel('SSIM')
fig.savefig('./plots/pretrained_mc.pdf', format="pdf", bbox_inches = 'tight')
```

```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('../styles_and_utils/graphstyle.mplstyle')

import sys
sys.path.insert(1, '../')
from styles_and_utils import *
```

```python
# pandas
import pandas as pd
def df2np(df, model):
    df = df.loc[df['Model'] == model, 'SSIM']
    return df.to_numpy().T

# Load brain results
brain_df = pd.read_excel('./data/brain_results.xlsx')

brain_unt_031M = df2np(brain_df, 'U-Net-31M')
brain_unt_124M = df2np(brain_df, 'U-Net-124M')
brain_vit_060M = df2np(brain_df, 'ViT-60M/10')
brain_vit_156M = df2np(brain_df, 'ViT-156M/16')
brain_pt_vit_156M = df2np(brain_df, 'PT-ViT-156M/16')



```

```python
w, h = plt.rcParams["figure.figsize"]
fig, ax = plt.subplots()

""" Model styles """
vits = ViT('s')
vitm = ViT('m')
vitl = ViT('l')

unts = Unet('s')
untm = Unet('m')
untl = Unet('l')

""" Plot brain """
brain_train_data = [2, 35, 70]
ax.plot(brain_train_data, brain_pt_vit_156M, '*--', markersize=vitl.ms, alpha=vitl.oc, color='#c93061', label=' PT-ViT-XL/16', zorder=5)
ax.plot(brain_train_data, brain_vit_156M, vitl.marker, markersize=vitl.ms, alpha=vitl.oc, color='#c93061', label = 'ViT-XL/16', zorder=4)
# ax.plot(brain_train_data, brain_vit_060M, vitl.marker, markersize=vitl.ms, alpha=vitl.oc, color=vitl.color, label=vitl.label+', P=10', zorder=3)
ax.plot(brain_train_data, brain_unt_124M, untl.marker, markersize=untl.ms, alpha=untl.oc, color=untl.color, label=untl.label, zorder=3)

# ax.set_title('ImageNet Pre-training')
ax.set_xticks(brain_train_data)
ax.set_xticklabels(['2k', '35k', '70k'])
ax.tick_params(axis='x', which='minor', bottom=False)
# ax.yaxis.set_ticks(np.arange(0.725, 0.745, 0.005))
ax.legend()

# Labels
ax.set_xlabel('Training data')
ax.set_ylabel('SSIM')

## Save
fig.savefig('./plots/pretrained_p16.pdf', format="pdf", bbox_inches = 'tight')
```
